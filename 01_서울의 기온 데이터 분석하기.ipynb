```python
import csv

min_temp = 999  # 최저 기온 값을 저장할 변수 -> 명확한 값을 위해 극적으로 초기값 설정
min_date = ""  # 최저 기온이 가장 낮았던 날짜를 저장할 변수

# 1) 파일을 연다
f = open("/content/seoul_2023(1).csv", "r", encoding="euc-kr")
# 2) 파일을 읽는다
data = csv.reader(f)

# 헤더(데이터 정보)저장하기
header = next(data)  # next()함수는 한줄 읽고, 커서(읽어드릴 위치) 한칸 이동

# 기온데이터 출력 -> 날짜의 최저 기온만 출력
for row in data:
    # 2-1) 결측치 데이터 처리 - > 명확한 결과를 위해 극적인 값으로 초기화
    if row[-2] == "":
        row[-2] = 999

    # 2-2 최저기온 실수형으로 형변함
    row[-2]=float(row[-2])

    # 3 최저기온과 날짜 찾기
    if min_temp > row[-2]:
        min_temp = row[-2]  # 최저기온
        min_date = row[0]  # 최저 기온의 날짜

# 4) 파일을 닫는다
f.close()

print(min_temp)
```

```python
import csv
import matplotlib.pyplot as plt

#1) 최고기온 담을 리스트 선언
result = []

min_temp = 999  # 최저 기온 값을 저장할 변수 -> 명확한 값을 위해 극적으로 초기값 설정
min_date = ""  # 최저 기온이 가장 낮았던 날짜를 저장할 변수

# 2) 파일을 연다
f = open("/content/seoul_2023(1).csv", "r", encoding="euc-kr")
# 2) 파일을 읽는다
data = csv.reader(f)

# 헤더(데이터 정보)저장하기3-1
header = next(data)  # next()함수는 한줄 읽고, 커서(읽어드릴 위치) 한칸 이동

#  3-2 최고기온 list  result 에 담기
for row in data:
    # 2-1) 결측치 데이터 처리 - > result 리스트에 담지 않겠다.(ignore)
    if row[-1] != '' :
      result.append(float(row[-1]))  #plot으로 형변환 해서 담기

f.close()

#5) 최고기온 데이터 시각화
plt.plot(result,"r")
plt.show()
```
