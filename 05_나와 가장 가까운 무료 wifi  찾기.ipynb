## 데이터 전처리

### 설치시설구분 컬럼 데이터 전처리

```python
df["설치시설구분"].unique()
```

```python
#'서민·복지시설',서민/복지시설,'서민복지시설',' 서민·복지시설'  ->서민·복지시설
```

```python
df.loc[df['설치시설구분'].str.contains(' 서민')] = '서민·복지시설' #위치.[값]이 있는 걸 찾아줌
#df["설치시설구분"].str.contains('서민')
df["설치시설구분"].unique()
```

```python
df.loc[df['설치시설구분'].str.contains('서민'), '설치시설구분'] = '서민·복지시설'
df["설치시설구분"].unique()
```

## 지역별 WIFI 설치현황 분석

### 설치시도별 설치시설구분과 서비스제공사 설치현황 집계표 만들기

```python
group_df = df.groupby(by = ['설치시도명', '설치시설구분','서비스제공사명']).size().reset_index(name='설치수')
group_df   # series
```

## 설치시도별 설치시설구분과 서비스제공사 설치현황 시각화



```python
group_df['서비스제공사명'].value_counts()
```

```python
px.histogram(group_df, x = '서비스제공사명', y = '설치수')
```

설치시도별 설치시설구분과 서비스제공사 설치현황

```python
px.histogram(group_df, x = '설치시설구분', y = '설치수')
```

```python
fig = px.pie(group_df, names = '설치시설구분', values = '설치수')
fig.update_traces(textposition = 'inside' , textinfo = 'percent + label')
```

```python
px.histogram(group_df, x = '설치시도명', y = '설치수')   # 시도명
```
