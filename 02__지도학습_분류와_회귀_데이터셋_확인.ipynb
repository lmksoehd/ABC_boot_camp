```python
#캘리포니아에서 평균집값이 housing_df['target'].max()와 같은 애들만 지도에 표시
```

```python
# 데이터 불러오기
housing = fetch_california_housing()
```

```python
# 데이터를 DataFrame으로 변환
housing_df = pd.DataFrame(housing.data, columns=housing.feature_names)
housing_df['target'] = housing.target
```

```python
# 'target' 열의 최댓값 구하기
max_target = housing_df['target'].max()
```

```python
housing_df[housing_df['target'] == max_target]
```

```python
# 최댓값과 같은 값을 가지는 샘플들을 지도에 표시

max_target_samples = housing_df[housing_df['target'] == max_target]

plt.scatter(max_target_samples['Longitude'], max_target_samples['Latitude'], s=50, c='red', label='최댓값 샘플')
plt.legend()
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('California Housing')
plt.show()


```

```python
import folium

max_target_samples = housing_df[housing_df['target'] == max_target]

for n in housing_df.index:
  price = (housing_df.loc[n, 'target'])*10000

  if housing_df.loc[n, 'target'] >= housing_df['target'].max():
    folium.Maker([housing_df.loc[n,'Latitude'], housing_df.loc[n, 'Longtitude']],
               popup=str(price)+'$',
               icon=folium.Icon(color='red', icon='fa-home', prefix='fa')).add_to(max_target_samples)

max_target_samples
```

```python
import folium

max_target_samples = housing_df[housing_df['target'] == max_target]

# 캘리포니아를 중심으로 하는 지도 생성
map = folium.Map(location=[36.7783, -119.4179], zoom_start=6)  # zoom_start

for n in max_target_samples.index:
    price = (max_target_samples.loc[n, 'target']) * 10000

    folium.Marker(
        location=[max_target_samples.loc[n, 'Latitude'], max_target_samples.loc[n, 'Longitude']],
        popup=str(price) + '$',
        icon=folium.Icon(color='red', icon='fa-home', prefix='fa')
    ).add_to(map)

# 지도를 HTML 파일로 저장
map
```
