```python
# BMI  
print("================================")
print("   BMI  ")
print("================================")
cm = float(input(' (cm) >>'))
kg = float(input(' (kg) >>'))
print("================================")
m = cm/100
m2=m*m
BMI = round( kg / m2 , 1) 

if BMI < 18.5:
    print("BMI \n{}\n".format(BMI))
elif 18.5 <= BMI< 22.9:
    print("BMI \n{}\n".format(BMI))
elif 23 <= BMI <24.9:
    print("BMI \n{}\n".format(BMI))
elif 25 <= BMI <29.9:
    print("BMI \n{}\n".format(BMI))
elif 30 <= BMI <34.9:
    print("BMI \n{}\n".format(BMI))
elif BMI >= 35:
    print("BMI \n{}\n".format(BMI))
print("================================")
```
