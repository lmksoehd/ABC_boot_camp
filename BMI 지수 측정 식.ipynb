```python
#1. float를 이용하여 키와 몸무게를 입력받음
#2. 키의 단위는 m, 몸무게는 kg 일 떼, BMI 지수는 신장제곱(M**2)/체중
#= 그렇다면 프로그램은 신장과 체중을 물어본 후 BMI 지수를 계산해서 내보낸다.
```

```python
# BMI 지수 계산식
print("================================")
print(" 체질량 지수 BMI를 확인해보세요 ")
print("================================")
cm = float(input('키를 입력하세요(cm) >>'))
kg = float(input('몸무게를 입력하세요(kg) >>'))
print("================================")
m = cm/100
m2=m*m
BMI = round( kg / m2 , 1) 

if BMI < 18.5:
    print("BMI결과 \n{}\n저체중입니다".format(BMI))
elif 18.5 <= BMI< 22.9:
    print("BMI결과 \n{}\n정상입니다".format(BMI))
elif 23 <= BMI <24.9:
    print("BMI결과 \n{}\n과체중입니다".format(BMI))
elif 25 <= BMI <29.9:
    print("BMI결과 \n{}\n경도비만입니다".format(BMI))
elif 30 <= BMI <34.9:
    print("BMI결과 \n{}\n중도비만입니다".format(BMI))
elif BMI >= 35:
    print("BMI결과 \n{}\n고도비만입니다".format(BMI))
print("================================")
```
