### 1 조건문을 이용해서 파생변수를 만들어라

#### 판정 합격 변수 만들기

# 통합 연비가 20이상이면 pass 그렇지 않으면 fail 을 부여하는 파생변수 만들기


```python
import numpy as np

mpg['test'] = np.where(mpg['total'] >= 20, 'pass', 'fail') # np.where(조건식, true, false)
mpg.head()
```

### 중첩 조건문 활용하기

####1. 연비 등급 변수 만들기

```python
#통합 연비 기준으로 A,B,C,D 등급 부여하는 grade 변수 만들기
''' a = 30/b=20/c는 그 외 '''
```

## 02-5 파생변수 추가

### 메소드로 파생변수 만들기


https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.assign.html

```python
#TOTAL (도시연비, 고속도로 연비/2)변수추가

mpg_df.assign(total = (mpg_df['cty']+ mpg_df['hwy'])/2)
```

#### 토탈변수의 평균값 넣기

```python
mpg_df.assign(total = (mpg_df['cty']+ mpg_df['hwy']),
              mean = (mpg_df['cty']+ mpg_df['hwy'])/2)
```

```python
mpg_df
```

```python
#### df.assign() 에 np.where 적용하기
```

```python
import numpy as np
mpg_df = mpg_df.assign(total=(mpg_df['cty'] + mpg_df['hwy']),
                       mean=(mpg_df['cty'] + mpg_df['hwy']) / 2)

mpg_df = mpg_df.assign(test=np.where(mpg_df['mean'] >= 20, 'pass', 'fail'))

print(mpg_df)
```
