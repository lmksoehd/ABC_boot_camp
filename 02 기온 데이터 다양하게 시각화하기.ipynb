```python
import csv
import matplotlib.pyplot as plt

#1) 01,08월 최고기온 담을 리스트 선언
aug = []
jan = []

min_temp = 999  # 최저 기온 값을 저장할 변수 -> 명확한 값을 위해 극적으로 초기값 설정
min_date = ""  # 최저 기온이 가장 낮았던 날짜를 저장할 변수

# 2) 파일을 연다
f = open("/content/seoul_2023(1).csv", "r", encoding="euc-kr")
# 2) 파일을 읽는다
data = csv.reader(f)

# 헤더(데이터 정보)저장하기3-1
header = next(data)  # next()함수는 한줄 읽고, 커서(읽어드릴 위치) 한칸 이동

#  3-2 최고기온 list  result 에 담기
for row in data:
    # 3-1) 결측치 데이터 처리 - > result 리스트에 담지 않겠다.(ignore)//8월만 빼고싶다 with 결측치 없이.
    if row[-1] != '' :
      #8월 데이터
      month = row[0].split('-')[1]
      if month =='01':
        jan.append(float(row[-1]))


      if month =='08':
        aug.append(float(row[-1]))



f.close()

plt.title("서울의 8월 최고 기온")



#5) 최고기온 데이터 시각화=> 히스토그램
plt.hist(jan, bins = 100, color = 'brown', label = "1월의 최고기온")
plt.hist(aug, bins = 100,color = 'green', label = "8월의 최고기온" )
plt.legend()
plt.show()
```

```python
import csv
import matplotlib.pyplot as plt

#1) 입력받은 월 최고기온 담을 리스트 선언
hi_temp = []
low_temp = []
avg_temp = []
press = int(input(" Press Month: "))


# open & read file
f = open("/content/seoul_2023(1).csv",'r', encoding = 'euc-kr')
data = csv.reader(f)

#sava header data
header = next(data)

#add 최고기온 in list
for row in data:
  #3-2) 결측치 데이터 처리

  month = int(row[0].split('-')[1])
  if row[-1] and row[-2] and row[-3] != '':
    if month == press:
      hi_temp.append(float(row[-1]))
      low_temp.append(float(row[-2]))
      avg_temp.append(float(row[-3]))

#close file
f.close()

#5) 히스토그램

plt.figure(dpi = 2000)
plt.title("서울의 %s월 최고, 평균, 최저 기온 히스토그램" %press)
plt.hist(hi_temp, bins=100, color = 'red', label = ('%s월 최고 기온' %press))
plt.hist(avg_temp, bins=100, color = 'green', label = ('%s월 최고 기온' %press))
plt.hist(low_temp, bins=100, color = 'blue', label = ('%s월 최고 기온' %press))
plt.legend()
plt.show()



'''
참고하세요

len을 써서 1자리수 입력 받아도 앞에 0을 붙혀줌
if len(press) == 1:
  press = press '0'+press


for row in data:
  month = row[0],split('-')[1]

  if month == press:
     if row[-1] !='' and row [-2] != row[-3] != '':  # 결측치 처리
      hi_temp.append(float(row[-1]))
      low_temp.append(float(row[-2]))
      avg_temp.append(float(row[-3]))

wanna go home
pigonhae digilmat


'''











```
