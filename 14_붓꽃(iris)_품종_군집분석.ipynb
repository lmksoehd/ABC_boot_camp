문제정의 : 붓꽃의 feature를 사용하여 군집 분석하고, 실제 품종과 군집 결과 비교하기

*   문제집 -> 데이터, 특성(feature), 독립변수(x) : 꽃잎, 꽃받침의 길이(cm) 4가지
*   정답 -> 클래스(class), 레이블(label), 타깃(target), 종속변수(y) : 붓꽃의 품종(setosa, versicolor, vriginica)

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

# 데이데프레임을 사용하여 데이터 분석 -> feature와 label의 연관성을 확인
iris_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)
iris_df.head()
```

군집 분석

```python
import numpy as np
from sklearn import cluster

X = iris_df.copy()
kmeans = cluster.KMeans(n_clusters=3, random_state=777)
kmeans.fit(X)

print(np.unique(kmeans.labels_))
cluster_label = kmeans.labels_
```

```python
# 예측 경과 데이터프레임에 추가
iris_df['Cluster'] = cluster_label
iris_df.tail()
```

```python
iris_df['target'] = iris_dataset['target']
```

```python
iris_dataset['target_names']
```

```python
iris_df
```

```python
iris_df['target_name'] = np.where(iris_df['target']==0, 'setosa', np.where(iris_df['target']==1, 'versicolor', 'virginica'))
iris_df
```

```python
iris_df['Cluster'].value_counts()
```

```python
# 데이데프레임을 사용하여 데이터 분석 -> feature와 label의 연관성을 확인
iris_dbm_df = pd.DataFrame(iris_dataset['data'], columns=iris_dataset.feature_names)

X = iris_dbm_df.copy()

dbm = cluster.DBSCAN(eps=0.8, min_samples=2)
dbm.fit(X)

print(np.unique(dbm.labels_))
dbm_cluster_label = dbm.labels_
```

```python
iris_dbm_df['Cluster'] = dbm_cluster_label
```

```python
iris_dbm_df['Cluster'].value_counts()
```

```python
iris_dbm_df['target'] = iris_dataset['target']
```

```python
iris_dbm_df
```
