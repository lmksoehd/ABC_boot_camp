```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
import koreanize_matplotlib

import warnings
warnings.filterwarnings('ignore')
```

```python
raw_welfare = pd.read_spss("/content/drive/MyDrive/새 폴더/Koweps_hpwc14_2019_beta2.sav")
raw_welfare
```

```python
welfare = raw_welfare.copy()
raw_welfare
```

```python
welfare.info()
```

```python
welfare.shape
```

```python
col_names = {'h14_g3': 'sex',
             'h14_g4': 'birth',
             'h14_g10': 'marriage_type',
             'h14_g11': 'religion',
             'p1402_8aq1': 'income',
             'h14_eco9': 'code_job',
             'h14_reg7': 'region'
            }



```

```python
welfare = welfare.rename(columns=col_names)
```

```python
welfare[['sex', 'birth', 'marriage_type', 'religion', 'income', 'code_job', 'region']]
```

### 1) 성별 데이터 확인하기

```python
print('성별데이터 타입 확인: ', welfare['sex'].dtypes)
print('성별데이터결측치확인: ', welfare['sex'].isna().sum())
print("성별 빈도 확인")
welfare['sex'].value_counts()
```

```python
welfare['sex'] = np.where(welfare['sex'] ==1, 'male','female')

print("성별 빈도 확인")
welfare['sex'].value_counts()
```

```python
sns.histplot(data = welfare, x = 'income')
plt.show()
```

```python
welfare.dropna(subset = ['income'], inplace=True)
print('성별데이터결측치확인: ', welfare['sex'].isna().sum())
```

### 5) 성별 월급 평균포 만들기

```python
# 성별 기준으로 월급의 평균표 만들기

sex_income = welfare.groupby('sex', as_index=False).agg(mean_income = ('income','mean'))
sex_income
```

### 06) 성별에 따른 월급 차이 분석

```python
sns.barplot(data = sex_income, x = 'sex', y = 'mean_income')
plt.show
```

#2) 나이와 월급의 관계 - 몇 살 때 월급을 가장 많이 벌까

### 1)나이데이터 확인하기

```python
print('나이 데이터 타입 확인 : ', welfare['income'].dtypes)
print('나이 데이터 결측지 확인 : ', welfare['income'].isna().sum())
print('나이 요약 통계량 확인')
welfare['income'].describe()
```

```python
print('나이 데이터 타입 확인 : ', welfare['birth'].dtypes)
print('나이 데이터 결측지 확인 : ', welfare['birth'].isna().sum())
print('나이 요약 통계량 확인')
welfare['birth'].describe()
```

```python
#그래프로 나이 분포도 확인

sns.histplot(data = welfare, x = 'birth')
plt.show()
```

### 02) 생년월일 전처리 => 파생 변수 생성-> 생년월일 활용하여 나이 데이터 만들기

```python
welfare = welfare.assign(age = 2019-welfare['birth']+1)
welfare['age']
```

```python
#그래프로 나이 분포도 확인

sns.histplot(data = welfare, x = 'age')
plt.show()
```

### 03) 나이에 따른 월급 평균표 만들기

```python
age_income = welfare.groupby('age', as_index = False).agg(mean_income = ('income','mean'))
age_income
```

###04) 나이와 월급의 관계 분석

```python
sns.lineplot(data=age_income, x='age', y='mean_income')
```

### 가장 월급을 많이 받는 나

```python
age_income['mean_income'].max()

```

```python
age_income.loc[age_income.idxmax(),['mean_income']]
```

### 05) 나이 및 성별 월급의 평균표

```python
sex_age_income = welfare.groupby('age', as_index = False).agg(mean_income = ('income','mean'))
sex_age_income
```

```python
sns.lineplot(data = sex_age_income, x = 'age', y = 'mean_income', hue = 'sex')
plt.show()
```
